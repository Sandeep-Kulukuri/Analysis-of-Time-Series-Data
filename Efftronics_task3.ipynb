# Feature Engineering

Adding Feature Date, Time, Day, Month, Year

```python
df=dict()
df['Date'] = [d.date() for d in df3['DateTime']]
df['Time'] = [d.time() for d in df3['DateTime']]
df['Day'] = [d.day_name() for d in df3['DateTime']]
df['Month'] = [d.month for d in df3['DateTime']]
df['Year'] = [d.year for d in df3['DateTime']]

df=pd.DataFrame(df)
#df.append(df3)
#Dropping DateTime column
for i in df3.keys():
    if i!='DateTime':
        df[i]=df3[i]

df
```

# Correlated Features

```python
correlated_features = set()
correlation_matrix = df3.corr()
for i in range(len(correlation_matrix .columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > 0.7:
            colname = correlation_matrix.columns[i]
            correlated_features.add(colname)
```

```python
print(correlated_features)
```

The Features Differences of Temperature and outside humidity are correlated i.e both features are dependent on each other.
Hence, these two features together imposes redundancy into the data. Therefore, further processsing of data requires only of this feature.
